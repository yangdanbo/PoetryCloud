```python
import tkinter
from tkinter import ttk
from tkinter import messagebox #弹窗库
from tkinter.messagebox import askyesno, askquestion
import matplotlib
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import numpy as np
import re
# encoding=utf-8
import jieba
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#唐诗
tang_poetry_list = []
#宋诗
song_poetry_list = []
#宋词
song_poem_list = []
#元曲
yuan_verse_list = []
#佛诗
buddhist_poetry_list = []

class PoetryWork(object):
    def __init__(self, num, title, author, content):
        self.num = num
        self.title = title
        self.author = author
        self.content = content
    def get_num(self):
        return self.num
    
    def get_author(self):
        return self.author
    
    def get_title(self):
        return self.title
    
    def get_content(self):
        return self.content
    
def get_screen_size(window):
    return window.winfo_screenwidth(), window.winfo_screenheight()
 
def get_window_size(window):
    return window.winfo_reqwidth(), window.winfo_reqheight()
 
def center_window(root,  width,  height):
    screenwidth = root.winfo_screenwidth()
    screenheight = root.winfo_screenheight()
    size = '%dx%d+%d+%d' % (width, height,  (screenwidth - width)/2,  (screenheight - height)/2)
    root.geometry(size)
    # root.resizable(0,0) 

# Exit GUI cleanly
def _quit():
    answer = askyesno(title='确认',
                          message='您确认要退出吗?')
    if answer:
        win.quit()
        win.destroy()

#加载唐诗三百首
def load_tang_shi():
    tang_poet_list = {}
    num = ''
    title = ''
    author = ''
    content = []
    pattern = r'\d{3}'
    with open('唐诗三百首.txt', 'r', encoding= 'utf-8') as f:
        for line in f.readlines():
            if line:
                if re.match(pattern, line):
                    #内容不空
                    if content:
                        tang_poetry_list.append(PoetryWork(num, title, author, content))
                    pos = line.index('：')
                    num = line[:3]
                    author = line[3:pos]
                    if author in tang_poet_list:
                        tang_poet_list[author] += 1
                    else:
                        tang_poet_list[author] = 1
                        
                    title = line[pos+1:]
                    content = []
                    #print('num:{},author:{},title:{}'.format(num, author, title))
                else:
                    if line.strip():
                        content.append(line.strip())
                    
    tang_poetry_list.append(PoetryWork(num, title, author, content))    
    tang_poet_list = dict(sorted(tang_poet_list.items(),key=lambda x:x[1], reverse=True))
    poet_list = [x for x in tang_poet_list.keys()]  
    poet_list.insert(0,'all')
    
    return poet_list

def strip_num(str):
    str = str.replace('①','').replace('②','').replace('③','').replace('④','').replace('⑤','')
    str = str.replace('⑥','').replace('⑦','').replace('⑧','').replace('⑨','').replace('⑩','')

    return str

#加载宋诗三百首
def load_song_shi():
    song_poet_list = {}
    num = ''
    title = ''
    author = ''
    content = []
    pattern = r'\d{3}'
    with open('宋诗三百首.txt', 'r', encoding= 'utf-8') as f:
        for line in f.readlines():
            line = line.strip()
            if line:
                if re.match(pattern, line):
                    #内容不空
                    if content:
                        song_poetry_list.append(PoetryWork(num, title, author, content))
                    pos = line.index('--')
                    num = line[:3]
                    title = strip_num(line[3:pos])
                    author = strip_num(line[pos+2:])
                    if author in song_poet_list:
                        song_poet_list[author] += 1
                    else:
                        song_poet_list[author] = 1
                    
                    content = []
                    #print('num:{},author:{},title:{}'.format(num, author, title))
                else:
                    if line.strip() and line[0] != '①':
                        content.append(strip_num(line.strip()))
                        
    song_poetry_list.append(PoetryWork(num, title, author, content))    
    song_poet_list = dict(sorted(song_poet_list.items(),key=lambda x:x[1], reverse=True))
    poet_list = [x for x in song_poet_list.keys()]  
    poet_list.insert(0,'all')

    return poet_list

#加载宋词三百首
def load_song_ci():
    song_poet_list = {}
    num = ''
    title = ''
    author = ''
    content = []
    pattern = r'\d{3}'
    with open('宋词三百首.txt', 'r', encoding= 'utf-8') as f:
        for line in f.readlines():
            line = line.strip()
            if line:
                if re.match(pattern, line):
                    #内容不空
                    if content:
                        song_poem_list.append(PoetryWork(num, title, author, content))
                    pos = line.index('--')
                    num = line[:3]
                    title = line[3:pos]
                    author = line[pos+2:]
                    if author in song_poet_list:
                        song_poet_list[author] += 1
                    else:
                        song_poet_list[author] = 1
                    
                    content = []
                    #print('num:{},author:{},title:{}'.format(num, author, title))
                else:
                    if line.strip() and ">>" not in line:
                        content.append(line.strip())
                        
    song_poem_list.append(PoetryWork(num, title, author, content))    
    song_poet_list = dict(sorted(song_poet_list.items(),key=lambda x:x[1], reverse=True))
    poet_list = [x for x in song_poet_list.keys()]  
    poet_list.insert(0,'all')

    return poet_list

#加载元曲三百首
def load_yuan_qu():
    yuan_poet_list = {}
    num = ''
    title = ''
    author = ''
    content = []
    pattern = r'\d{3}'
    with open('元曲三百首.txt', 'r', encoding= 'utf-8') as f:
        for line in f.readlines():
            line = line.strip()
            if line:
                if re.match(pattern, line):
                    #内容不空
                    if content:
                        yuan_verse_list.append(PoetryWork(num, title, author, content))
                    pos = line.index('--')
                    num = line[:3]
                    title = line[3:pos].strip()
                    author = line[pos+2:]
                    if author in yuan_poet_list:
                        yuan_poet_list[author] += 1
                    else:
                        yuan_poet_list[author] = 1
                    
                    content = []
                    #print('num:{},author:{},title:{}'.format(num, author, title))
                else:
                    if line.strip() and ">>" not in line:
                        content.append(line.strip())
                        
    yuan_verse_list.append(PoetryWork(num, title, author, content))    
    yuan_poet_list = dict(sorted(yuan_poet_list.items(),key=lambda x:x[1], reverse=True))
    poet_list = [x for x in yuan_poet_list.keys()]  
    poet_list.insert(0,'all')

    return poet_list

#加载佛诗三百首
def load_buddhist_poerty():
    buddhist_poet_list = {}
    num = ''
    title = ''
    author = ''
    content = []
    pattern = r'\d{3}'
    with open('佛诗三百首.txt', 'r', encoding= 'utf-8') as f:
        for line in f.readlines()[330:]: #跳过前331行目录
            line = line.strip()
            if line:
                if re.match(pattern, line):
                    #内容不空
                    if content:
                        buddhist_poetry_list.append(PoetryWork(num, title, author, content))
                    pos = line.index('--')
                    num = line[:3]
                    title = line[3:pos].strip()
                    author = line[pos+2:]
                    if author in buddhist_poet_list:
                        buddhist_poet_list[author] += 1
                    else:
                        buddhist_poet_list[author] = 1
                    
                    content = []
                    #print('num:{},author:{},title:{}'.format(num, author, title))
                else:
                    if line.strip() and ">>" not in line:
                        content.append(line.strip())
                        
    buddhist_poetry_list.append(PoetryWork(num, title, author, content))    
    buddhist_poet_list = dict(sorted(buddhist_poet_list.items(),key=lambda x:x[1], reverse=True))
    poet_list = [x for x in buddhist_poet_list.keys()]  
    poet_list.insert(0,'all')

    return poet_list

def changeWork(*args):   #处理事件，*args表示可变参数
    #print(workList.get()) #打印选中的值
    if workList.get() == '宋诗三百首':
        if not song_poetry_list:
            poetList["values"]=load_song_shi()
            poetList.current(0)
    if workList.get() == '宋词三百首':
        if not song_poem_list:
            poetList["values"]=load_song_ci()
            poetList.current(0)
    if workList.get() == '元曲三百首':
        if not yuan_verse_list:
            poetList["values"]=load_yuan_qu()
            poetList.current(0)    
    if workList.get() == '佛诗三百首':
        if not buddhist_poetry_list:
            poetList["values"]=load_buddhist_poerty()
            poetList.current(0)   

def changePoet(*args):
    cur_poetry_list = []
    cur_author = poetList.get()
        
def generateCloud(*args):
    """清除原有图表，生成新的图表"""
    global fig,canvas
    fig.clear()

    ax = fig.add_subplot(111)
    content = []
    cur_author = poetList.get()
    if workList.get() == '唐诗三百首':
        for work in tang_poetry_list:
            if cur_author == 'all' or cur_author == work.get_author():
                content.extend(work.get_content())
    elif workList.get() == '宋诗三百首':
        for work in song_poetry_list:
            if cur_author == 'all' or cur_author == work.get_author():
                content.extend(work.get_content())
    elif workList.get() == '宋词三百首':
        for work in song_poem_list:
            if cur_author == 'all' or cur_author == work.get_author():
                content.extend(work.get_content())
    elif workList.get() == '元曲三百首':
        for work in yuan_verse_list:
            if cur_author == 'all' or cur_author == work.get_author():
                content.extend(work.get_content())   
    else: #佛诗三百首
        for work in buddhist_poetry_list:
            if cur_author == 'all' or cur_author == work.get_author():
                content.extend(work.get_content())  
                
    res = jieba.lcut(" ".join(content))  # 中文分词
    text = " ".join(res)  # 用空格连接所有的词
    mask = np.array(Image.open("chinamap.png"))  # 指定词云图效果
    # 创建词云对象
    wc = WordCloud(width=800, height=600, mask=mask).generate(text)
    ax.imshow(wc) # 显示词云图
    ax.axis("off")
 
    wc.to_file("{}_{}_wordcloud.png".format(workList.get(), cur_author))  # 保存成图片
    
    canvas.draw()
    
# Create instance
win = tkinter.Tk() #构造窗体  

# Add a title       
win.title("诗词云图")  
win.iconbitmap("poetry.ico")

#设定600*480居中,
center_window(win, 600, 480)

selectCondition = ttk.LabelFrame(win, text="", width=640)
selectCondition.pack(side='top')

#选择作品
curRow = 0
lblWorks =  ttk.Label(selectCondition,  text="作品：")
lblWorks.grid(column=0, row=curRow, sticky='W', pady=4)

selectedWork = tkinter.StringVar()#窗体自带的文本，新建一个值
workList = ttk.Combobox(selectCondition, textvariable=selectedWork) #初始化
workList["values"]=("唐诗三百首","宋诗三百首","宋词三百首","元曲三百首","佛诗三百首")
workList.current(0)  #选择第一个
workList.bind("<<ComboboxSelected>>", changeWork)  #绑定事件,(下拉列表框被选中时，绑定go()函数)
workList.grid(column=1, row=curRow, sticky='W', padx=4, pady=4)

btnCloud = ttk.Button(selectCondition,  text='生成词云', width=8, command=generateCloud)
btnCloud.grid(column=2, row=curRow, sticky='W', pady=4)

#选择作者
curRow = 1
lblPoet = ttk.Label(selectCondition,  text="作者：")
lblPoet.grid(column=0, row=curRow, sticky='W', pady=4)

selectedPoet = tkinter.StringVar()#窗体自带的文本，新建一个值
poetList = ttk.Combobox(selectCondition, textvariable=selectedPoet) #初始化
#poetList["values"]=("李白","杜甫","白居易","王维","李商隐","杜牧")
#poetList.current(0)  #选择第一个
poetList.bind("<<ComboboxSelected>>", changePoet)  #绑定事件,(下拉列表框被选中时，绑定go()函数)
poetList.grid(column=1, row=curRow, sticky='W', padx=4, pady=4)

btnExit = ttk.Button(selectCondition, text='退出', width=8, command=_quit)
btnExit.grid(column=2, row=curRow, sticky='W', pady=4)

curRow = 2
fig = Figure(figsize=(5, 4), dpi=100)
canvas = FigureCanvasTkAgg(fig, master=win)  # A tk.DrawingArea.
canvas.get_tk_widget().pack(side=tkinter.BOTTOM, fill=tkinter.BOTH, expand=1)

poetList["values"]=load_tang_shi()
poetList.current(0)

# 生成并显示图表
generateCloud()

win.mainloop() #进入消息循环
```
